# 团结引擎

本题考察Unity引擎及Mod的安装与卸载。如果你选择解包并且不想和我一样不慎注册了*团结引擎*发现打不开项目，可以先搜搜exe详细信息中的版本号：



### Flag 1

强大的Unity插件系统[BepInEx](https://github.com/BepInEx/BepInEx)，支撑着多么惊人的力量。

本题我使用了[UnityExplorer](https://github.com/originalnicodr/CinematicUnityExplorer)，内置了变速齿轮，组件浏览器，FreeCam等解本题所有需要的功能。

可以使用变速齿轮加速，或者直接使用FreeCam开自由视角。Flag 1在旗杆后面的角落里。如果你找不到这一个Flag，说明你应该多玩⚪⚪。



### Flag 2
